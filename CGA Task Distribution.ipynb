```python
# Ac Nice
new_name = sheet.Name.str.split() # Split into a list
lasts = []
firsts = []
middles = []

for key,name in new_name.items():
    if len(name) >= 3:
        lasts.append(str(name[0]).strip())
        firsts.append(str(name[1]).strip())
        middles.append(str(name[2]).strip())
    elif len(name) >= 2:
        lasts.append(str(name[0]).strip())
        firsts.append(str(name[1]).strip())
        middles.append('NaN')

sheet['Lastname'] = lasts
sheet['Firstname'] = firsts
sheet['Middlename'] = middles

# Reindex the columns
indices = ['Timestamp', 'Score','Lastname','Firstname','Middlename',
           'Name','Track','Team','Responsibility 1','Responsibility 2',
           'Responsibility 3','Topic 1','Topic 2','Topic 3','Day 1','Day 2',
           'Crosscheck']
sheet = sheet.reindex(columns=indices)
sheet.head()
```
